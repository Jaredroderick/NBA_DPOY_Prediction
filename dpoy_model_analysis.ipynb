Using the nba_api, import the data for leaguedashplayerstats and leaguedash defend.

```python
# List the seasons that we are dating back to. I chose 2013 because that is when the NBA started using player-tracking data
seasons = ['2013-14', '2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23', '2023-24', '2024-25']
all_seasons_data = []

# Loop through each season and collect all of the players' data
for s in seasons:
    total_player_stats = leaguedashplayerstats.LeagueDashPlayerStats(season=s)
    total_player_stats_df = total_player_stats.get_data_frames()[0]
    total_player_stats_df['SEASON_ID'] = s
    all_seasons_data.append(total_player_stats_df)
    print(f"Loaded {len(total_player_stats_df)} rows for {s}.")
```
